```python
from datasets import load_dataset

dataset = load_dataset("HuggingFaceM4/COCO")
```

```python
from tqdm import tqdm
for_csv = {
    "imgid":[],
    "sentid":[],
    "cocoid":[],
    "raw":[]

}

for example in tqdm(dataset["train"]):
  for_csv["imgid"].append(example["imgid"])
  for_csv["sentid"].append(example["sentences"]["sentid"])
  for_csv["cocoid"].append(example["cocoid"])
  for_csv["raw"].append(example["sentences"]["raw"])
```

```python
import pandas as pd
df = pd.Dataframe(for_csv)
df.to_csv("COCO_train_sentences.csv")
```

```python
for_csv = {
    "imgid":[],
    "sentid":[],
    "cocoid":[],
    "raw":[]

}

for example in tqdm(dataset["test"]):
  for_csv["imgid"].append(example["imgid"])
  for_csv["sentid"].append(example["sentences"]["sentid"])
  for_csv["cocoid"].append(example["cocoid"])
  for_csv["raw"].append(example["sentences"]["raw"])
```

```python
import pandas as pd
df = pd.DataFrame(for_csv)
df.to_csv("COCO_test_sentences.csv")
```
